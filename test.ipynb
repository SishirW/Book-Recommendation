```python
import numpy as np
import pandas as pd
```

```python
books=pd.read_csv("Books.csv")
ratings=pd.read_csv("Ratings.csv")
```

```python
# Popularity
ratings_with_name= ratings.merge(books,on='ISBN')
ratings_with_name.head(10)
```

```python
num_ratings=ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
num_ratings.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
```

```python
num_ratings
```

```python
avg_rating = ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating.size
```

```python
popular_df = num_ratings.merge(avg_rating,on='Book-Title')
popular_df.size
```

```python
popular_df = popular_df[popular_df['num_ratings']>=100].sort_values('avg_rating',ascending=False)
```

```python
popular_df.head(10)
```

```python


x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
```

```python
valid_users= x[x].index
```

```python
valid_users
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(valid_users)]
```

```python
filtered_rating
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
final_ratings
```

```python
pt = final_ratings.pivot_table(index='ISBN',columns='User-ID',values='Book-Rating')
# pt = final.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
#pt=pd.read_csv('pt.csv')
pt.to_csv('popular.csv')
```

```python
# pt.fillna(0,inplace=True)
# pt
#pt.to_csv('popularity_table.csv')
#pt
pt=pd.read_csv('pt.csv')
pt = pt.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores
```

```python
def recommend(book_name):
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```
